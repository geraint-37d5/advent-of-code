# Advent of Code: Day 10

```julia
using Base.Iterators
using DataStructures
using Match
using Underscores

#f = "ExampleInput.txt"
f = "SolutionInput.txt"
I = readlines(f)
```

## Part 1

```julia
C = Dict(
  ']' => '[',
  ')' => '(',
  '}' => '{',
  '>' => '<'
)

p(L) = begin
  o = nothing
  S = []
  for c ∈ L
    if c ∈ ['[','(','{','<']
      push!(S, c)
    elseif C[c] == last(S)
      pop!(S)
    else
      o = c
      break
    end
  end
  o
end

SC = Dict(
  ')' => 3,
  ']' => 57,
  '}' => 1197,
  '>' => 25137
)

@_ I                     |>
   map(p, __)            |>
   filter(_!=nothing,__) |>
   map(SC[_],__)         |>
   sum
```

## Part 2

```julia
O = Dict(
  '[' => ']',
  '(' => ')',
  '{' => '}',
  '<' => '>'
)

SC2 = Dict(
  ')' => 1,
  ']' => 2,
  '}' => 3,
  '>' => 4
)

p2(L) = begin
  o = nothing
  S = []
  for c ∈ L
    if c ∈ ['[','(','{','<']
      push!(S, c)
    elseif C[c] == last(S)
      pop!(S)
    end
  end
  @_ S                                    |>
     map(O[_], __)                        |>
     reverse                              |>
     reduce((_1 * 5) + SC2[_2],__;init=0)
end

@_ I                             |>
   map((p(_),_), __)             |>
   filter(_[1]==nothing,__)      |>
   map(_[2],__)                  |>
   map(p2,__)                    |>
   sort                          |>
   __[ceil(Int, length(__) / 2)]
   
```
